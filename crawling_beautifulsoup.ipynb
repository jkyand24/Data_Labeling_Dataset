```python
import requests
from bs4 import BeautifulSoup
import os

# request 보내 response 받기

query = "apple"

url = f"https://www.google.com/search?q={query}&source=lnms&tbm=isch"

header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

response = requests.get(url, headers=header)

# response를 BeautifulSoup 객체애게 전달하여, 모든 img tag 얻기

soup = BeautifulSoup(response.text, "html.parser")

img_tags = soup.find_all("img")

# 모든 img tag 내의 url? 모아놓기

urls = []

for img in img_tags:
    try:
        urls.append(img["data-src"])
    except KeyError:
        try:
            urls.append(img["data-iurl"])
        except KeyError:
            try:
                urls.append(img['src'])
            except KeyError:
                pass

# 모든 url로부터 image를 얻어 새 폴더에 저장

os.makedirs(query, exist_ok=True)

for i, url in enumerate(urls):
    try:
        img_data = requests.get(url, headers=header).content
        file_path = os.path.join(query, f"{query}_{i}.jpg")
        with open(file_path, "wb") as f:
            f.write(img_data)
        if os.path.getsize(file_path) == 0:
            os.remove(file_path)
    except:
        pass
```
